# Models

### Split the data into Training and Test offical sets

#### Benchmark split

```python
X_train_Bench = feature_data.loc[feature_data['data_set'] == 0]
Y_train_Bench = feature_data['PHQ8_Binary'].loc[feature_data['data_set'] == 0]
X_train_Bench.drop("PHQ8_Binary", axis=1, inplace=True)
X_train_Bench.drop(["PHQ8_Score", "Gender", "data_set", "patient_id"], axis=1, inplace=True)

X_test_Bench = feature_data.loc[feature_data['data_set'] == 2]
Y_test_Bench = feature_data['PHQ8_Binary'].loc[feature_data['data_set'] == 2]
X_test_Bench.drop("PHQ8_Binary", axis=1, inplace=True)
X_test_Bench.drop(["PHQ8_Score", "Gender", "data_set", "patient_id"], axis=1, inplace=True)
```

#### Random Data Split

```python
X_features = feature_data.drop(["PHQ8_Binary","PHQ8_Score", "Gender", "data_set", "patient_id"], axis=1)
Y_MDD_binary = feature_data['PHQ8_Binary']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_features, Y_MDD_binary, test_size = 0.20, random_state = 1)
```

#### Down Sampling 

```python
Train_dataset = pd.concat([X_train, Y_train], axis=1)
```

```python
Train_dataset
```

```python
X_dataset_majority = Train_dataset[Train_dataset["PHQ8_Binary"]==0]
X_dataset_minority = Train_dataset[Train_dataset["PHQ8_Binary"]==1]
```

```python
from sklearn.utils import resample
X_dataset_majority_downsampled = resample(X_dataset_majority, 
                                 replace=False,    
                                 n_samples=len(X_dataset_minority))
```

```python
df_up_down_sampled = pd.concat([X_dataset_majority_downsampled, X_dataset_minority])
Y_train_dsample = df_up_down_sampled['PHQ8_Binary']
X_train_dsample = df_up_down_sampled.drop("PHQ8_Binary", axis=1, inplace=False)
```

#### Down samping before random split

```python
dataset_majority = feature_data[feature_data["PHQ8_Binary"]==0]
dataset_minority = feature_data[feature_data["PHQ8_Binary"]==1]
```

```python
from sklearn.utils import resample
dataset_majority_downsampled = resample(dataset_majority, 
                                 replace=False,    
                                 n_samples=len(dataset_minority), random_state =1)
```

```python
dataset_downsampled = pd.concat([dataset_majority_downsampled, dataset_minority])
```

```python
X_train_DS_RS = dataset_downsampled.loc[dataset_downsampled['data_set'] == 0]
Y_train_DS_RS = dataset_downsampled['PHQ8_Binary'].loc[dataset_downsampled['data_set'] == 0]
X_train_DS_RS.drop("PHQ8_Binary", axis=1, inplace=True)
X_train_DS_RS.drop(["PHQ8_Score", "Gender", "data_set", "patient_id"], axis=1, inplace=True)

X_test_DS_RS = dataset_downsampled.loc[dataset_downsampled['data_set'] == 2]
Y_test_DS_RS = dataset_downsampled['PHQ8_Binary'].loc[dataset_downsampled['data_set'] == 2]
X_test_DS_RS.drop("PHQ8_Binary", axis=1, inplace=True)
X_test_DS_RS.drop(["PHQ8_Score", "Gender", "data_set", "patient_id"], axis=1, inplace=True)
```

```python
dataset_downsampled
```

### SVM Linear

```python
from sklearn import svm
SVM_Linear_Model = svm.SVC(kernel='linear')
SVM_Linear_Model.fit(X_train_DS_RS, Y_train_DS_RS)
```

```python
Y_pred_SVM_linear = SVM_Linear_Model.predict(X_test_DS_RS)
```

```python
from sklearn import metrics
print("The accuracy is ", metrics.accuracy_score(Y_test_DS_RS, Y_pred_SVM_linear))
print("The precision is",metrics.precision_score(Y_test_DS_RS, Y_pred_SVM_linear))
print("The recall is",metrics.recall_score(Y_test_DS_RS, Y_pred_SVM_linear))
```

### SVM Gausian 

```python
from sklearn import svm
SVM_Gausian_Model = svm.SVC(kernel='rbf')
SVM_Gausian_Model.fit(X_train, Y_train)
```

```python
Y_pred_SVM_Gausian = SVM_Gausian_Model.predict(X_test)
```

```python
from sklearn import metrics
print("The accuracy is ", metrics.accuracy_score(Y_test, Y_pred_SVM_Gausian))
print("The precision is",metrics.precision_score(Y_test, Y_pred_SVM_Gausian))
print("The recall is",metrics.recall_score(Y_test, Y_pred_SVM_Gausian))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, Y_pred_SVM_Gausian))
print(classification_report(Y_test, Y_pred_SVM_Gausian))
```

### MLR

```python
from sklearn.linear_model import LogisticRegression
MLR_Model = LogisticRegression(random_state=16)
MLR_Model.fit(X_train, Y_train)
Y_pred_MLR = MLR_Model.predict(X_test)
```

```python
print("The accuracy is ", metrics.accuracy_score(Y_test, Y_pred_MLR))
print("The precision is",metrics.precision_score(Y_test, Y_pred_MLR))
print("The recall is",metrics.recall_score(Y_test, Y_pred_MLR))
```
